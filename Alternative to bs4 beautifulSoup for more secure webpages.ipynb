### Suppose we want to automatically go to next page

```python
list_review_df = []

pageNum = 1
while True:
    try:
        print(f"page {pageNum}")
        
        time.sleep(1)
        
        df_review_page = pd.DataFrame()
        
        #------------- retrieve the data
        
        # get names
        usernames = driver.find_elements_by_xpath("//div[@id='cm_cr-review_list']/div[@data-hook='review']//div[@class='a-profile-content']/span[@class='a-profile-name']")
        usernames_list = [username.get_attribute('innerHTML') for username in usernames]
        print(usernames_list)
        
        # get titles
        titles = driver.find_elements_by_xpath("//div[@id='cm_cr-review_list']/div[@data-hook='review']//a[@data-hook='review-title']")
        title_list = [title.text for title in titles]
        title_seri = pd.Series(title_list)
        
        #------------- store the data into the page dataframe
        df_review_page['Username'] = pd.Series(usernames_list)
        df_review_page['Title'] = pd.Series(title_list)
        print(df_review_page)
        
        #------------- adding this dataframe into a list in order to make a global dataframe 
        list_review_df.append(df_review_page)
        
        #------------- going to next page

        nextPageButton = driver.find_element_by_xpath("//ul[@class='a-pagination']/li[2]")
        
        if 'a-disabled' in nextPageButton.get_attribute('class'):
            print('We have reached the last page!')
            break
            
        nextPageButton.click()
        print(f"------------------ Done with {pageNum} ------------------\n")
        pageNum += 1
        time.sleep(1)
        print(f"Done with {pageNum}")
        
    except Exception as e:
        print(e)
        break
```
